```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## realtion is non linear

X=7*np.random.rand(100 ,1)-np.random.rand(100,1) - 3

Y=0.5*(X**2) + 1.5*X + 2*np.random.rand(100,1) +7 ## qudratice eq

#polynomial degree=2, multiple regression
```

```python
plt.scatter(X,Y,color='g')
plt.title('Scatter Plot')
plt.ylabel('Y')
plt.xlabel('X')
```

```python
## train n test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
## scaling

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
## simple linear

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
## score

sc=lr.score(X_train,Y_train)
print('score',sc)
##r2 score

from sklearn.metrics import r2_score
r2=r2_score(Y_test,lr.predict(X_test))
print('r2 score',r2)
```

```python
## visulaistion
plt.plot(X_train,lr.predict(X_train),color='r')
plt.scatter(X_train,Y_train,color='g')
```

```python
## not fitted well
```

```python
## polynomial regression

##increase degree

from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=4,include_bias=False) # bias -> constant too c + bx + ax^2 , c=1
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
## NOW LINEAR REGRESSION

lr=LinearRegression()
lr.fit(X_train_poly,Y_train)
r2=r2_score(Y_test,lr.predict(X_test_poly))
print('r2 score',r2)
```

```python
## visulaistion
plt.scatter(X_train,lr.predict(X_train_poly),color='r')
plt.scatter(X_train,Y_train,color='g')
```
